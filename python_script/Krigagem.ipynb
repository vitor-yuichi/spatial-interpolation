**Análise Descritiva dos dados**

Os pontos estão nas coordenadas EPSG:4674 - SIRGAS 2000 - Zona 23 (América do Sul). Para utilizá-los na interpolação de qualquer método, é necessário a conversão para um sistema de unidades em metros.


```python
from pyproj import Proj, transform, Transformer, CRS

def convert_sirgas_to_utm(longitude, latitude, sirgas_zone):
    # Define SIRGAS 2000 projection
    sirgas_proj = CRS.from_epsg(4674)

    # Convert SIRGAS coordinates to UTM
    utm_proj = CRS(f'EPSG:{32700 + sirgas_zone}')

    transformer = Transformer.from_crs(sirgas_proj, utm_proj, always_xy=True)
    easting, northing = transformer.transform(longitude, latitude)

    return easting, northing
```

**Convertendo do EPSG 4674 para SIRGAS 2000 zona 23.**

```python
longitude_utm = []
latitude_utm = []

for longs, lat in zip(df.longitude, df.latitude):
  longitude_utm.append(convert_sirgas_to_utm(longs, lat, 23)[0])
  latitude_utm.append(convert_sirgas_to_utm(longs, lat, 23)[1])
```

```python
df['longitude_utm'] = longitude_utm
df['latitude_utm'] = latitude_utm
```

## Resultados do variograma experimental:

O variograma experimental foi executado com todos os pluviômetros existente na área de estudo.


**A linha verde**: É o modelo teórico de ajuste (Inicialmente esférico);

**Os pontos azuis**: são os valores da semivariânca representado em (mm²) (Eixo y). E representa a distribuição do variograma;


**A semivariância de Matheron é uma medida de dispersão ou variabilidade de um conjunto de dados geoespaciais abaixo de um determinado limiar. Ela é uma extensão da ideia de semivariância, mas incorpora uma abordagem anisotrópica, levando em consideração a direção e a distância entre os pontos.;**

**A quantidade de lags**: Foi ajustado para o total de 10 lags;

**O eixo x**: Está em metros;


Visualmente podemos observar que não há a tendência geral do aumento lag com o aumento proporcional da semivariância. Em geral o variograma experimental representado pelos pontos azuis indica baixa continuidade espacial entre os dados;

Foram construídos variogramas com 80%, 60% e até 40% das amostras. Para todos estes casos temos baixa correlação de aumento do lag e o aumento da semivariância;





```python
# df[['xcoord', 'ycoord']]
d = df.iloc[0:23, :]
V1 = Variogram(d[['latitude_utm', 'longitude_utm']].values, d.pluvio_trimester.values, normalize=False, estimator = 'matheron')
V1.n_lags = 3
V1.plot(show=False)
plt.style.use('ggplot')
```

**Calculando-se o variograma experimental com os diversos modelos teóricos de ajuste e cálculo do Erro Quadrático Médio (RMSE)**

Para determinar o melhor modelo de ajuste, podemos selecionar o modelo com menor desvio RMSE.

**Entretanto** nenhum modelo teórico ajusta-se na distribuição.


```python
fig, _a = plt.subplots(2,3, figsize=(18, 10), sharex=True, sharey=True)
axes = _a.flatten()
for i, model in enumerate(('spherical', 'exponential', 'gaussian', 'matern', 'cubic')):
    V1.model = model
    V1.plot(axes=axes[i], hist=False, show=False)
    axes[i].set_title('Model: %s; RMSE: %.2f' % (model, V1.rmse))
```
